```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("ecommerce_data.csv")
```

```python
label = ["is_mobile","n_products_viewed","visit_duration","is_returning_visitor","time_of_day","user_action"]
```

```python
def getMax(label):
    a = 0
    for i in label:
        if i >= a:
            a = i
    return a
```

```python
maxdata = getMax(data["is_mobile"])
```

```python
maxdata = getMax(data["is_mobile"])
data["is_mobile"] = data["is_mobile"]/maxdata
```

```python
for i in label:
    maxdata = getMax(data[i])
    data[i] = data[i]/maxdata
```
