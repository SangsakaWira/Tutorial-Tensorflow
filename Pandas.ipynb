```python
import numpy as np
import pandas as pd
```

```python
data = np.random.randn(100)
```

```python
data
```

```python
labels = ["A","B","C","D","E","F","G","H","I","J"]
df = pd.DataFrame(data.reshape(10,10)**2,columns=labels)
```

```python
def normalize(df):
    result = df.copy()
    for feature_name in df.columns:
        max_value = df[feature_name].max()
        min_value = df[feature_name].min()
        result[feature_name] = (df[feature_name] - min_value) / (max_value - min_value)
    return result
```

```python
df_normal = normalize(df) 
```

```python
df_normal["new"] = np.random.rand(10)
```

```python
df_normal.drop("new",axis=1,inplace=True)
```

```python
df_normal
```

```python
df_normal["A"] = df_normal["B"]
```

```python
df_normal[["A","C"]] = df_normal[["B","B"]]
```
